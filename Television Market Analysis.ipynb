**Importing Libraries and Datasets of Television**

```python
import altair as alt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/SOUBHIK MANDAL/Desktop/jupyter notebook/Television Market Analysis/TV_Final.csv')
df
```

```python
colors = ["#0000FF","#00FF00","#FFFF00","#FF7F00","#FF0000","#4361EE"]
palette = sns.color_palette(palette = colors)

sns.palplot(palette, size =1)
plt.show()
```

**Types of Television availables in the market**

```python
df['Brand'].groupby(df['Resolution']).count().sort_values(ascending=False)
```

```python
labels = 'Ultra HD LED', 'QLED Ultra HD', 'HD LED', 'Full HD LED','HD Plasma'
sizes = [399,229,220,60,4]
fig, ax = plt.subplots(figsize=(10,6))
fig.set_facecolor('white')
ax.pie(sizes, labels=labels, colors=["#0000FF","#00FF00","#FFFF00","#FF7F00","#FF0000"],autopct='%1.1f%%', startangle=60,textprops={'color':'black','weight':'bold','fontsize':12.5})
ax.axis('equal')
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(10,7))
sns.countplot(y="Brand", hue="Resolution", data=df, palette=colors,order=df["Brand"].value_counts().iloc[:10].index)
```

**Brands of Television availables in the market**

```python
df['Brand'].groupby(df['Brand']).count().sort_values(ascending=False).iloc[:10]
```

**Rating of the Television Brands availables in the market**

```python
round(df.groupby('Brand')['Rating'].mean().sort_values(ascending=False),1)
```

```python
list1 = ["SONY","OnePlus","Haier","Vu","TOSHIBA","MOTOROLA","Mi","Compaq","Onida","Thomson",
        "KODAK","LG","realme","SAMSUNG","Nokia","Panasonic","Sansui","Micromax","PHILIPS","Samsung","JVC"]

list2 = ["Hisense","Thomson","Onix","KODAK","Dektron","iFFALCON by TCL","MarQ by Flipkart","Nokia","Coocaa","Blaupunkt","Weston","TCL",
        "Infinix","Sanyo","Intex","IMPEX","Croma","HUIDI","CloudWalker","Lloyd", 
        "Candes","Detel","LumX"]
series1 = df["Brand"].isin(list1)
series2 = df["Brand"].isin(list2)
df_f1= df[series1]
df_f2= df[series2]
fig, axs = plt.subplots(ncols=2,figsize=(20,5))
ax1 = sns.stripplot(x="Rating", y="Selling Price", data=df_f1,hue="Brand", palette=colors,ax=axs[0])
ax2 = sns.stripplot(x="Rating", y="Selling Price", data=df_f2,hue="Brand", palette=colors,ax=axs[1])
```

**Average selling price by Brand, Resolution & Operating system**

```python
df.groupby('Brand')['Selling Price'].mean().sort_values(ascending=False).apply(np.ceil)
```

```python
alt.Chart(df).mark_point(filled=True,size=40).encode(
    x='Original Price',
    y='Selling Price',
    color='Brand',
    tooltip=['Brand', 'Selling Price', 'Original Price']
).interactive()
```
